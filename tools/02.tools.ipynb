```python
from typing import Any

from langchain_community.tools import StructuredTool
from pydantic import Field, BaseModel
import warnings
warnings.filterwarnings('ignore')
```

```python

class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="the first number to add")
    b: int = Field(required=True, description="the first number to add")
```

```python
def multiply_fnc(a: int,b: int) -> int:
    return a*b
```

```python
struc_tool = StructuredTool.from_function(
    args_schema=MultiplyInput,
    func=multiply_fnc,
    description="multiply two numbers",
    name="multipy"
)
```

```python
final = struc_tool.invoke({'a':4, 'b':5})
```

```python
final
```

```python
print(final)
print(struc_tool.name)
print(struc_tool.description)
print(struc_tool.args)
```

```python
from langchain.tools import BaseTool
```

```python
from typing import Type
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"
    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```
